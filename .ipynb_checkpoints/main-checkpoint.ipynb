```python
import util_mnist_reader
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import to_categorical
from keras.layers import MaxPooling2D
import keras.callbacks
from keras.layers import Conv2D
from keras.layers import Flatten
```

```python
def sigmoid(x):
    return(1/(1+np.exp(-x)))

def softmax(x):
    a = np.exp(x)
    return a/np.sum(a,axis=0)

#plot loss for keras
class plotloss(keras.callbacks.Callback):
    
    def on_train_begin(self,logs={}):
        self.losses = []

    def on_epoch_end(self, epoch, logs={}):
        self.losses.append(logs.get('loss')) #append loss in every epoch
```

```python
def task1(X_train,X_test,Yn_train,Y_test):

    lr = 0.75
    h_neurons = 50
    o_neurons = 10
    b_ih = np.zeros((h_neurons,1))
    b_ho = np.zeros((o_neurons,1))
    epochs = 2000
    w_ih = np.random.randn(X_train.shape[0],h_neurons)
    w_ho = np.random.randn(h_neurons,o_neurons)
    m = X_train.shape[1]

    trackloss = []

    for epoch in range(epochs):

        #Feed foward
        zih = np.dot(w_ih.T,X_train) + b_ih
        aih = sigmoid(zih)
        zho = np.dot(w_ho.T,aih) + b_ho
        aho = softmax(zho)

        #calculate Loss
        #loss = calc_loss(Yn_train,aho,m)
        Loss_sum = np.sum(np.multiply(Yn_train,np.log(aho)))
        Loss = -(1/m)*Loss_sum
        #print("Epoch ",epoch," Loss :",Loss)
        trackloss.append(np.squeeze(Loss))
        
        #Back propagation
        d_zho =  aho - Yn_train
        d_who = (1/m)*np.dot(aih,d_zho.T)
        d_bho = (1/m)*np.sum(d_zho,axis=1,keepdims=True)
        
        d_aih = np.dot(w_ho,d_zho)
        a = aih*(1-aih) #variable for reference
        d_zih = d_aih*a
        d_wih = (1/m)*np.dot(X_train,d_zih.T)
        d_bih = (1/m)*np.sum(d_zih, axis=1,keepdims = True)

        w_ih = w_ih - (lr*d_wih)
        w_ho = w_ho - (lr*d_who)
        b_ho = b_ho - (lr*d_bho)
        b_ih = b_ih - (lr*d_bih)

    print("\nLoss:\t",Loss)
    plt.plot(trackloss,label = 'task1 loss')
    plt.legend()
    #plt.show()  

    #TESTING
    z1 = np.dot(w_ih.T,X_test) + b_ih
    a1 = sigmoid(z1)
    z2 = np.dot(w_ho.T,a1) + b_ho
    a2 = softmax(z2)

    # print(a2.shape)

    predictions = np.argmax(a2,axis=0)
 
    accuracy = accuracy_score(Y_test,predictions)
    print("\nAccuracy:\t",accuracy*100,"%")
    conf_matrix = confusion_matrix(Y_test,predictions)
    print("\nConfusion matrix:\n",conf_matrix)
    #print(classification_report(Y_test,predictions))
    return
```

```python
def task2(x,y,a,b):

    #TASK 2
    loss_graph = plotloss() #object of the class to initialize
    
        
    #create a model
    m = Sequential([
        Dense(60,activation='sigmoid', input_shape=(784,)),
        Dense(60,activation='sigmoid'),
        Dense(60,activation='sigmoid'),
        Dense(10,activation='softmax'),
    ])
    
    #compile the model
    m.compile(
        optimizer='adam',
        loss='categorical_crossentropy',
        metrics=['accuracy'],
    )

    #train the model over the training dataset
    m.fit(
        x,
        to_categorical(y),
        epochs=50,
        batch_size=30,
        callbacks=[loss_graph], #call the object to append loss 
    )
    
    #evaluate the model against the test data
    eval = m.evaluate(
        a,
        to_categorical(b)
    )
    
    #     plt.plot(loss_graph.losses)
    
    #     print("\nLoss:\t",eval[0])
    #     print("\nAccuracy:\t",eval[1]*100,"%")

    predictions = m.predict(a)

    pred = np.argmax(predictions,axis=1)
    accuracy = accuracy_score(b,pred)
    conf_matrix = confusion_matrix(b,pred)
    #     print("\nConfusion matrix:\n",conf_matrix)


    return eval,conf_matrix,loss_graph.losses,accuracy
```

```python
def task3(x,y,a,b):

    loss_graph = plotloss() #to plot loss

    filters = 8
    fsize = 3
    pool = 2
    print(x.shape)
    print(a.shape)
    # x = np.expand_dims(x, axis=2)
    # a = np.expand_dims(a, axis=2)
    x = X_train.reshape(60000,28,28)
    a = X_test.reshape(10000,28,28)
    x = np.expand_dims(x, axis=3) # get it to shape 60000,28,28,1
    a = np.expand_dims(a, axis=3) # get it to shape 10000,28,28,1
    # print(x.shape)
    # print(a.shape)

    #build model of the CNN
    m = Sequential([
        Conv2D(filters,fsize,input_shape=(28,28,1)),
        MaxPooling2D(pool_size=pool),
        Flatten(),
        Dense(60,activation='sigmoid'),
        Dense(60,activation='sigmoid'),
        Dense(10,activation='softmax'),
    ])

    #compile the model
    m.compile(
        optimizer='adam',
        loss='categorical_crossentropy',
        metrics=['accuracy']

    )
    
    #train the model over the training dataset
    m.fit(
        x,
        to_categorical(y),
        epochs=15,
        batch_size=60,
        callbacks=[loss_graph],
    )
    
    #evaluate the model against the test data
    eval = m.evaluate(
        a,
        to_categorical(b)
    )


    #     plt.plot(loss_graph.losses)
    #     plt.show()

    #     print("Output of keras:\n")
    #     print("\nLoss:\t",eval[0])
    #     print("\nAccuracy:\t",eval[1]*100,"%")

    predictions = m.predict(a)

    pred = np.argmax(predictions,axis=1)
    accuracy = accuracy_score(b,pred)
    #print("\nAccuracy:\t",accuracy*100,"%")
    conf_matrix = confusion_matrix(b,pred)
    #     print("\nConfusion matrix:\n",conf_matrix)

    return eval,conf_matrix,loss_graph.losses,accuracy
```

```python
X_train, Y_train = util_mnist_reader.load_mnist('data/fashion', kind='train')
X_test, Y_test = util_mnist_reader.load_mnist('data/fashion', kind='t10k')
Yn_train = np.zeros([Y_train.shape[0],10])
Yn_test = np.zeros([Y_test.shape[0],10])

#bring labels to the correct form - (60000,10)
for i in range(Y_train.shape[0]):
    j = Y_train[i]
    for k in range(10):
        if(k==j):
            Yn_train[i][k] = 1


Xn_train = X_train.T
Xn_test = X_test.T
Yn_train = Yn_train.T

print("\n\n##################################################################################\n\n")
print("\n\t\t\tTASK1 - Neural Network with one hidden layer\n\n\n")

task1(Xn_train,Xn_test,Yn_train,Y_test)
print("\n\n##################################################################################\n\n")
print("\n\t\t\t\tEND OF TASK1\n")
```

```python
print("\n\n##################################################################################\n\n")
print("\n\t\t\tTASK2 - Multi-Layer Neural Network with keras\n\n\n")
eval,conf_matrix,l,acc1 = task2(X_train,Y_train,X_test,Y_test)

```

```python
print("\nLoss:\t",eval[0])
print("\nAccuracy:\t",acc1*100,"%")
print("\nConfusion matrix:\n",conf_matrix)

plt.plot(l,label = 'task2 loss')
plt.legend()

print("\n\n##################################################################################\n\n")
print("\n\t\t\t\tEND OF TASK2\n\n\n")
print(" Loss graph:")
```
